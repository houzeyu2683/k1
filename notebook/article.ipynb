```python
import pandas
import os
from sklearn import preprocessing
```

```python
root = "/".join(os.getcwd().split("/")[:-1])
path = os.path.join(root, 'resource/kaggle/csv/articles.csv')
```

```python
table = pandas.read_csv(path, dtype=str)
table['detail_desc'] = table['detail_desc'].fillna("<NA>")
```

```python
encoder = preprocessing.LabelEncoder()
head = 10
key = "article_id_code"
encoder.fit(table['article_id'])
table[key] = encoder.transform(table['article_id']) + head
```
